```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.display_functions import display
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import warnings
warnings.filterwarnings("ignore")
```

```python
df_agg = pd.read_csv(r"Aggregation.txt", sep="\s", header=None, engine='python')
df_agg = df_agg.rename(columns={0:'feature1', 1:'feature2', 2:'cluster'})
display(df_agg.head())
    
df_flm = pd.read_csv(r"flame.txt", sep="\s", header=None, engine='python')
df_flm = df_flm.rename(columns={0:'feature1', 1:'feature2', 2:'cluster'})
display(df_flm.head())

df_jain = pd.read_csv(r"jain.txt", sep="\s", header=None, engine='python')
df_jain = df_jain.rename(columns={0:'feature1', 1:'feature2', 2:'cluster'})
display(df_jain.head())

df_sprl = pd.read_csv(r"spiral.txt", sep="\s", header=None, engine='python')
df_sprl = df_sprl.rename(columns={0:'feature1', 1:'feature2', 2:'cluster'})
display(df_sprl.head())
```

```python
plt.title("Aggregation data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_agg[['feature1']], y = df_agg[['feature2']])
plt.show()
```

```python
plt.title("flame data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_flm[['feature1']], y = df_flm[['feature2']])
plt.show()
```

```python
plt.title("jain data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_jain[['feature1']], y = df_jain[['feature2']])
plt.show()
```

```python
plt.title("spiral data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_sprl[['feature1']], y = df_sprl[['feature2']])
plt.show()
```

# k-means

```python
x_agg = df_agg[['feature1', 'feature2']].values
cluster_count = int(max(df_agg[['cluster']].values))
detected_cluster = KMeans(n_clusters=cluster_count).fit_predict(x_agg)
df_agg['new_cluster'] = detected_cluster
display(df_agg)
```

```python
plt.title("Aggregation data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_agg[['feature1']], y = df_agg[['feature2']], c=df_agg[['new_cluster']].values)
plt.show()
```

```python
x_flm = df_flm[['feature1', 'feature2']].values
cluster_count = int(max(df_flm[['cluster']].values))
detected_cluster = KMeans(n_clusters=cluster_count).fit_predict(x_flm)
df_flm['new_cluster'] = detected_cluster
display(df_flm)
```

```python
plt.title("flame data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_flm[['feature1']], y = df_flm[['feature2']], c=df_flm[['new_cluster']].values)
plt.show()
```

```python
x_jain = df_jain[['feature1', 'feature2']].values
cluster_count = int(max(df_jain[['cluster']].values))
detected_cluster = KMeans(n_clusters=cluster_count).fit_predict(x_jain)
df_jain['new_cluster'] = detected_cluster
display(df_jain)
```

```python
plt.title("jain data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.scatter(x = df_jain[['feature1']], y = df_jain[['feature2']], c=df_jain[['new_cluster']].values)
plt.show()
```

```python
x_sprl = df_sprl[['feature1', 'feature2']].values
cluster_count = int(max(df_sprl[['cluster']].values))
detected_cluster = KMeans(n_clusters=cluster_count).fit_predict(x_sprl)
df_sprl['new_cluster'] = detected_cluster
display(df_sprl)
```
